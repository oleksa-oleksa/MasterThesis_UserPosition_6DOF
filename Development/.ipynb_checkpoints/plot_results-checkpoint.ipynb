```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

 
# take data
data = pd.read_csv("results/gpu_jobs_results_flipped_lstm_MSE.csv")
 
df = pd.DataFrame(data, columns=["MSE_pos", "hidden_dim", "epochs", "batch_size", 'layers', 'dropout'])
 
# plot the dataframe
df.plot(x="MSE_pos", y=["hidden_dim", "epochs", "batch_size"], kind="bar", figsize=(9, 8))
 
# print bar graph
plt.show()
```

```python
# plot the dataframe
df.plot(x="MSE_pos", y="hidden_dim", kind="bar", figsize=(9, 8))
 
# print bar graph
plt.show()
```

```python
#plotting boxplot 
plt.boxplot(df) 

#x and y-axis labels 
plt.xlabel('hidden_dim') 
plt.ylabel('MSE_pos') 

#plot title 
plt.title('Analysing hidden_dim') 

#save and display 
#plt.savefig('C:\\Users\\Dell\\Desktop\\AV Plotting images\\matplotlib_plotting_9.png',dpi=300,bbox_inches='tight') 
plt.show();
```

```python
import toml
import os
path = "data/interpolated/1556.csv"
dataset_hmd = pd.read_csv(path)
cfg = toml.load('config.toml')
dataset_lengh_sec = 600
start = 400
end = 500
ts_start = int(start/0.005)
ts_end = int(end/0.005)
        
ts = np.arange(0, dataset_lengh_sec + cfg['dt'], cfg['dt'])
ts_comp = ts[ts_start:ts_end]
fig, (ax) = plt.subplots(1, 1, figsize=(18, 4), sharex=True)

# Plot position in Quaternions
ax.plot(ts_comp, dataset_hmd.loc[ts_start:ts_end-1, 'y'], label='y', linestyle='--', color='tab:orange')
ax.set_xlabel('seconds')
ax.set_ylabel('degrees')
ax.set_xlim(start, end)
plt.ylim(-0.5, 0.5)
ax.legend(loc='upper left')
ax.yaxis.grid(which='major', linestyle='dotted', linewidth=1)
#ax.xaxis.set_major_locator(MultipleLocator(10))

trace_id = os.path.splitext(os.path.basename(path))[0]
dest = os.path.join('results/plotter', f"Fig-{trace_id}-y_position.pdf")
fig.savefig(dest)
```

```python
import toml
import os
path = "data/interpolated/1556.csv"
dataset_hmd = pd.read_csv(path)
cfg = toml.load('config.toml')
dataset_lengh_sec = 600
start = 0
end = 600
ts_start = int(start/0.005)
ts_end = int(end/0.005)
        
ts = np.arange(0, dataset_lengh_sec + cfg['dt'], cfg['dt'])
ts_comp = ts[ts_start:ts_end]
fig, (ax) = plt.subplots(1, 1, figsize=(18, 4), sharex=True)

# Plot position in Quaternions
ax.plot(ts_comp, dataset_hmd.loc[ts_start:ts_end-1, 'z'], label='z', linestyle='--', color='tab:orange')
ax.set_xlabel('seconds')
ax.set_ylabel('degrees')
ax.set_xlim(start, end)
plt.ylim(-12, 12)
ax.legend(loc='upper left')
ax.yaxis.grid(which='major', linestyle='dotted', linewidth=1)
#ax.xaxis.set_major_locator(MultipleLocator(10))
plt.axhline(y=0.0, color='black', linestyle='-')


trace_id = os.path.splitext(os.path.basename(path))[0]
dest = os.path.join('results/plotter', f"Fig-{trace_id}-z_position_interpolated.pdf")
fig.savefig(dest)
```

```python
import toml
import os
path = "data/normalized_min-max/1556.csv"
dataset_hmd = pd.read_csv(path)
cfg = toml.load('config.toml')
dataset_lengh_sec = 600
start = 0
end = 600
ts_start = int(start/0.005)
ts_end = int(end/0.005)
        
ts = np.arange(0, dataset_lengh_sec + cfg['dt'], cfg['dt'])
ts_comp = ts[ts_start:ts_end]
fig, (ax) = plt.subplots(1, 1, figsize=(18, 4), sharex=True)

# Plot position in Quaternions
ax.plot(ts_comp, dataset_hmd.loc[ts_start:ts_end-1, 'z'], label='z', linestyle='--', color='tab:orange')
ax.set_xlabel('seconds')
ax.set_ylabel('degrees')
ax.set_xlim(start, end)
plt.ylim(-0.5, 1.5)
ax.legend(loc='upper left')
ax.yaxis.grid(which='major', linestyle='dotted', linewidth=1)
#ax.xaxis.set_major_locator(MultipleLocator(10))
plt.axhline(y=0.0, color='black', linestyle='-')
plt.axhline(y=1.0, color='black', linestyle='-')



trace_id = os.path.splitext(os.path.basename(path))[0]
dest = os.path.join('results/plotter', f"Fig-{trace_id}-z_position_min-max.pdf")
fig.savefig(dest)
```
