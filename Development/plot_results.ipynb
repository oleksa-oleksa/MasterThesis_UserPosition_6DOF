```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

 
# take data
data = pd.read_csv("results/gpu_jobs_results_flipped_lstm_MSE.csv")
 
df = pd.DataFrame(data, columns=["MSE_pos", "hidden_dim", "epochs", "batch_size", 'layers', 'dropout'])
 
# plot the dataframe
df.plot(x="MSE_pos", y=["hidden_dim", "epochs", "batch_size"], kind="bar", figsize=(9, 8))
 
# print bar graph
plt.show()
```

```python
# plot the dataframe
df.plot(x="MSE_pos", y="hidden_dim", kind="bar", figsize=(9, 8))
 
# print bar graph
plt.show()
```

```python
#plotting boxplot 
plt.boxplot(df) 

#x and y-axis labels 
plt.xlabel('hidden_dim') 
plt.ylabel('MSE_pos') 

#plot title 
plt.title('Analysing hidden_dim') 

#save and display 
#plt.savefig('C:\\Users\\Dell\\Desktop\\AV Plotting images\\matplotlib_plotting_9.png',dpi=300,bbox_inches='tight') 
plt.show();
```
