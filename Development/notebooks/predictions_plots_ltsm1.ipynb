```python
import toml
import matplotlib.pyplot as plt
cfg = toml.load('../UserPrediction6DOF/tools/config.toml')
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import os

dest = '../results/plotter/pred_vs_real/'
model = 'lstm1'

targets = pd.read_csv('../results/targets/targets.csv')
base = pd.read_csv('../results/predictions/baseline/baseline_predictions.csv', names=["x", "y", "z", "qx", "qy", "qz", "qw"])

dataset_lengh_sec = targets.shape[0]
ts = np.arange(0, dataset_lengh_sec + cfg['dt'], cfg['dt'])

predictions = pd.read_csv('../results/predictions/lstm1/MAE_pos_0.013_lstm1_hid512_batch128_epochs500_LR0.0001_every30_epochs_with_WD1e-12_for_LAT100_08.08_083011.csv',
                          names=["x", "y", "z", "qx", "qy", "qz", "qw"])
```

```python
print(base.head(25))
```

```python
compare1 = 'x'
comp1 = plt.figure()
plt.plot(targets.loc[:len(ts)-1, compare1],label='target_X')
plt.plot(predictions.loc[:len(ts)-1, compare1],label='pred_X')
plt.plot(base.loc[:len(ts)-1, compare1],label='base_X')
comp1.set_figwidth(20)
comp1.set_figheight(10)
plt.legend()
plt.show()
dest1 = os.path.join(dest, f"{model}-{compare1}.pdf")
comp1.savefig(dest1)
```
