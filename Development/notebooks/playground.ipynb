```python
import toml
import os
path = "data/interpolated/1556.csv"
dataset_hmd = pd.read_csv(path)
cfg = toml.load('config.toml')
dataset_lengh_sec = 600
start = 400
end = 500
ts_start = int(start/0.005)
ts_end = int(end/0.005)
        
ts = np.arange(0, dataset_lengh_sec + cfg['dt'], cfg['dt'])
ts_comp = ts[ts_start:ts_end]
fig, (ax) = plt.subplots(1, 1, figsize=(18, 4), sharex=True)

# Plot position in Quaternions
ax.plot(ts_comp, dataset_hmd.loc[ts_start:ts_end-1, 'y'], label='y', linestyle='--', color='tab:orange')
ax.set_xlabel('seconds')
ax.set_ylabel('degrees')
ax.set_xlim(start, end)
plt.ylim(-0.5, 0.5)
ax.legend(loc='upper left')
ax.yaxis.grid(which='major', linestyle='dotted', linewidth=1)
#ax.xaxis.set_major_locator(MultipleLocator(10))

trace_id = os.path.splitext(os.path.basename(path))[0]
dest = os.path.join('results/plotter', f"Fig-{trace_id}-y_position.pdf")
fig.savefig(dest)
```
